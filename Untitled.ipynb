Winsorize (cap) the upper tail of only the “heavy-tailed” features.
This avoids outliers dragging K-Means centroids.
cap level: Use 99th percentile (good default).

Cap only those that commonly have extreme values:
- total_purchases
- recency_days
- avg_gap_days
- max_gap_days
- customer_age_days
- recency_to_avg_gap (sometimes)
- (optional) log_total_spend (usually already fine)

Do not cap bounded ratios unless needed:
- share_A is already [0,1]
- active_ratio is [0,1]

```python
import numpy as np

model_cols = [
    'total_purchases',
    'recency_days',
    'avg_gap_days',
    'max_gap_days',
    'recency_to_avg_gap',
    'log_total_spend',
    'log_avg_spend',
    'customer_age_days',
    'active_ratio',
    'share_A'
]

# Features to cap (exclude bounded ratios unless you want)
cap_cols = [
    'total_purchases',
    'recency_days',
    'avg_gap_days',
    'max_gap_days',
    'recency_to_avg_gap',
    'customer_age_days',
    'log_total_spend',   # optional; keep if you want extra safety
    'log_avg_spend'      # optional
]

q = 0.99  # cap at 99th percentile

user_model = user_agg[model_cols].copy()

for col in cap_cols:
    upper = user_model[col].quantile(q)
    user_model[col] = user_model[col].clip(upper=upper)
```

```python
outlier_report = []
for col in cap_cols:
    p95 = user_model[col].quantile(0.95)
    p99 = user_model[col].quantile(0.99)
    mx  = user_model[col].max()
    outlier_report.append((col, p95, p99, mx))

outlier_report[:5]  # or print it nicely

# If max ≫ p99, you had outliers — and capping was useful.
```
